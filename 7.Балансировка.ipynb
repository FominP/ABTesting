Объявим нужные нам функции

```python
def calc_strat_mean(values, labels, weights):
    df = pd.concat([values, labels], axis=1)
    df.columns = ['values', 'labels']
    strat_mean = df.groupby('labels')['values'].mean()
    return (strat_mean * weights).sum()

def calc_strat_var(values, labels, weights):
    df = pd.concat([values, labels], axis=1)
    df.columns = ['values', 'labels']
    strat_var = df.groupby('labels')['values'].var()
    return (strat_var * weights).sum()

def transform_cuped(control, test, control_before, test_before):
    num = np.cov(control, control_before)[0, 1] + np.cov(test, test_before)[0, 1]
    den = np.var(control_before) + np.var(test_before)
    theta = num / den
    control_cup = control - theta * control_before
    test_cup = test - theta * test_before
    return control_cup, test_cup

def absolute_ttest(
    control,
    test,
    control_before=None,
    test_before=None,
    control_labels=None,
    test_labels=None,
    weights=None,
    ci_level=0.05
):
    # CUPED
    if control_before is not None:
        control, test = transform_cuped(
          control=control,
          test=test,
          control_before=control_before,
          test_before=test_before,
        )
    # стратификация
    if control_labels is not None:
        mean_control = calc_strat_mean(control, control_labels, weights)
        mean_test = calc_strat_mean(test, test_labels, weights)
        var_mean_control = calc_strat_var(control, control_labels, weights) / len(control)
        var_mean_test = calc_strat_var(test, test_labels, weights) / len(control)
    else:
        mean_control = np.mean(control)
        mean_test = np.mean(test)
        var_mean_control  = np.var(control) / len(control)
        var_mean_test  = np.var(test) / len(test)
    difference_mean = mean_test - mean_control
    difference_mean_var = var_mean_control + var_mean_test
    difference_distribution = norm(loc=difference_mean, scale=np.sqrt(difference_mean_var))
    left_bound, right_bound = difference_distribution.ppf([ci_level / 2, 1 - ci_level / 2])
    pvalue = 2 * min(difference_distribution.cdf(0), difference_distribution.sf(0))
    effect = difference_mean
    return experiment_result(pvalue, effect, left_bound, right_bound)

def do_synthetic_aa(
    data,
    data_before=None,
    data_labels=None,
    stratify_strategy=None,
    test_size=0.5,
    test_func=absolute_ttest,
    alpha=0.05,
    n_iter=10_000,
):
    """stratify_strategy = None | 'split' | 'mean' | 'both'
    функция принимает np.array или pd.Series,
    после чего проводит синтетические АА-тесты по след. алгоритму:
    - разделить на тест и контроль (доля теста = test_size),
    - провести стат. тест (test_func - absolute_ttest по умолчанию) и записать p value,
    - повторить шаги n_iter (10 тысяч раз по умолчанию),
    - рассчитать % ошибок 1 рода и дов. интервал (относительно уровня alpha),
    - вывести результат и вернуть массив p value (если включен флаг return_pvalues)"""
    p_values = []
    for i in tqdm(range(n_iter)):
        if stratify_strategy in ('split', 'both'):
            control, test = train_test_split(
                data,
                stratify=data_labels,
                test_size=0.5
            )
        else:
            control, test = train_test_split(data, test_size=0.5)
        control_before, test_before = None, None
        if data_before is not None:
            control_before = data_before.loc[control.index]
            test_before = data_before.loc[test.index]
        control_labels, test_labels, weights = None, None, None
        if stratify_strategy in ('mean', 'both'):
            control_labels = data_labels.loc[control.index]
            test_labels = data_labels.loc[test.index]
            weights = data_labels.value_counts() / len(data_labels)
        p_values.append(test_func(
            control=control,
            test=test,
            control_before=control_before,
            test_before=test_before,
            control_labels=control_labels,
            test_labels=test_labels,
            weights=weights,
        ).pvalue)
    p_values = np.array(p_values)
    fte = (p_values <= alpha).sum()
    left_real_level, right_real_level = proportion_confint(
        count=fte,
        nobs=n_iter,
        alpha=alpha,
        method='wilson',
    )
    significance = fte / n_iter
    return synthetic_tests_result(left_real_level, significance, right_real_level)

def do_synthetic_ab(
    data,
    data_before=None,
    data_labels=None,
    stratify_strategy=None,
    potential_effect=0.1,
    test_size=0.5,
    test_func=absolute_ttest,
    alpha=0.05,
    n_iter=10_000,
):
    """stratify_strategy = None | 'split' | 'mean' | 'both'
    функция принимает np.array или pd.Series,
    после чего проводит синтетические АB-тесты по след. алгоритму:
    - разделить на тест и контроль (доля теста = test_size),
    - построить ДИ для разницы средних и записать размер эффекта и записать их,
    - повторить предыдущие шаги n_iter (10 тысяч раз по умолчанию),
    - рассчитать % ошибок 2 рода и дов. интервал (относительно уровня alpha),
    - вернуть значения из предыдущего пункта"""
    left_bounds = []
    right_bounds = []
    absolut_effect = data.mean() * potential_effect
    control_labels, test_labels, weights = None, None, None
    control_before, test_before = None, None
    for i in tqdm(range(n_iter)):
        if stratify_strategy in ('split', 'both'):
            control, test = train_test_split(
                data,
                stratify=data_labels,
                test_size=0.5
            )
        else:
            control, test = train_test_split(data, test_size=0.5)
        test *= 1 + potential_effect
        if data_before is not None:
            control_before = data_before.loc[control.index]
            test_before = data_before.loc[test.index]
        if stratify_strategy in ('mean', 'both'):
            control_labels = data_labels.loc[control.index]
            test_labels = data_labels.loc[test.index]
            weights = data_labels.value_counts() / len(data_labels)
        _, _, left_bound, right_bound = test_func(
            control=control,
            test=test,
            control_before=control_before,
            test_before=test_before,
            control_labels=control_labels,
            test_labels=test_labels,
            weights=weights,
        )
        left_bounds.append(left_bound)
        right_bounds.append(right_bound)
    left_bounds, right_bounds = np.array(left_bounds), np.array(right_bounds)
    ste = ((left_bounds > absolut_effect) | (right_bounds < absolut_effect)).sum()
    left_real_level, right_real_level = proportion_confint(
        count=ste,
        nobs=n_iter,
        alpha=alpha,
        method='wilson',
    )
    significance = ste / n_iter
    return synthetic_tests_result(left_real_level, significance, right_real_level)
```

```python
for metric in metrics:
    test = df_two_periods[df_two_periods['is_test']][f"{metric}_t2"]
    control = df_two_periods[~df_two_periods['is_test']][f"{metric}_t2"]
    control_before = df_two_periods.loc[control.index][f"{metric}_t1"]
    test_before = df_two_periods.loc[test.index][f"{metric}_t1"]
    activity_type = df_two_periods.loc[control.index]['activity_type']

    # Абсолютный t-тест
    pvalue, effect, ci_low, ci_high = absolute_ttest(control, test, ci_level=0.05)
    results['metric'].append(metric)
    results['method'].append('absolute_ttest')
    results['p_value'].append(pvalue)
    results['effect'].append(effect)
    results['ci_low'].append(ci_low)
    results['ci_high'].append(ci_high)

    # Абсолютный t-тест с пост-стратификацией
    data_labels=df_two_periods['activity_type']
    control_labels = data_labels.loc[control.index]
    test_labels = data_labels.loc[test.index]
    weights = data_labels.value_counts() / len(data_labels)
    pvalue, effect, ci_low, ci_high = absolute_ttest(
        control=control,
        test=test,
        control_labels=control_labels,
        test_labels=test_labels,
        weights=weights,
        ci_level=0.05)
    results['metric'].append(metric)
    results['method'].append('absolute_ttest_stratified')
    results['p_value'].append(pvalue)
    results['effect'].append(effect)
    results['ci_low'].append(ci_low)
    results['ci_high'].append(ci_high)

    # Абсолютный t-тест с CUPED
    pvalue, effect, ci_low, ci_high = absolute_ttest(
        control=control,
        test=test,
        control_before=control_before,
        test_before=test_before,
        ci_level=0.05)
    results['metric'].append(metric)
    results['method'].append('absolute_ttest_cuped')
    results['p_value'].append(pvalue)
    results['effect'].append(effect)
    results['ci_low'].append(ci_low)
    results['ci_high'].append(ci_high)

    # Абсолютный t-тест с CUPED и пост-стратификацией
    pvalue, effect, ci_low, ci_high = absolute_ttest(
        control=control,
        test=test,
        control_before=control_before,
        test_before=test_before,
        control_labels=control_labels,
        test_labels=test_labels,
        weights=weights,
        ci_level=0.05)
    results['metric'].append(metric)
    results['method'].append('absolute_ttest_cuped_stratified')
    results['p_value'].append(pvalue)
    results['effect'].append(effect)
    results['ci_low'].append(ci_low)
    results['ci_high'].append(ci_high)


results_df = pd.DataFrame(results)

results_df['ci_width'] = results_df['ci_high'] - results_df['ci_low']

ci_width_comparison = results_df.pivot(index='metric', columns='method', values='ci_width')
ci_width_comparison['absolute_ttest_stratified_reduction'] = (ci_width_comparison['absolute_ttest'] - ci_width_comparison['absolute_ttest_stratified']) / ci_width_comparison['absolute_ttest'] * 100
ci_width_comparison['absolute_ttest_cuped_reduction'] = (ci_width_comparison['absolute_ttest'] - ci_width_comparison['absolute_ttest_cuped']) / ci_width_comparison['absolute_ttest'] * 100
ci_width_comparison['absolute_ttest_cuped_stratified_reduction'] = (ci_width_comparison['absolute_ttest'] - ci_width_comparison['absolute_ttest_cuped_stratified']) / ci_width_comparison['absolute_ttest'] * 100

results_df
```
